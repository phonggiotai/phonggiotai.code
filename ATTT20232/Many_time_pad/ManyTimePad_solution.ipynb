<div >Được giải dựa trên nguyên lí (a XOR k) XOR (b XOR k) = a XOR b (1) <br>
Các bản mã a' b' được tạo ra bằng phép xor với k<br>
k là key mã hoá, khi được sử dụng nhiều hơn 1 lần, kẻ tấn công có khả năng tận dụng nguyên lí trên để tìm ra chuỗi a xor b ban đầu, loại bỏ vai trò của key mã hoá <br>
Sau đó phán đoán dựa trên tận suất xuất hiện các chữ, chủ đề của thông điệp để dần tìm ra ab.</div> Phương pháp giải mã này dựa vào ngữ nghĩa của thông điệp khá nhiều.

Trước tiên ta thực hiện đoán từ xuất hiện nhiều nhất ở đầu câu là 'the' tương đương với đoạn bản mã '32510b' xuất hiện 5 lần, tuy nhiên theo nguyên lí (1) ta không cần quan tâm đến đoạn bản mã này, mà thực hiện XOR 1 bản mã khác với bản mã chứa 'the':


```python
#xor 2 chuỗi hex, cần chuyển chuỗi hex về số thập phân rồi xor 2 số thập phân, sau đó chuyển lại thành chuỗi hex
def xor_two_string(string1, string2):
    #chuyển chuỗi hex về số thập phân
    num1 = int(string1, 16)
    num2 = int(string2, 16)
    #xor 2 số thập phân
    xor_num = num1 ^ num2
    #chuyển số thập phân về chuỗi hex
    xor_string = hex(xor_num)[2:]
    #nếu chuỗi hex có độ dài lẻ thì thêm 0 ở đầu
    if len(xor_string) % 2 != 0:
        xor_string = '0' + xor_string
    return xor_string

def hex_to_ascii(hex_string):
    return bytes.fromhex(hex_string).decode('utf-8')
```
